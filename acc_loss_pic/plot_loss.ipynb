```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exp_name = 'metrics_aa_ls'
```

```python
df = pd.read_csv('vit_logs/'+exp_name+'.csv')
val_loss = df['val_loss'].dropna().reset_index(drop=True)
val_acc = df['val_acc'].dropna().reset_index(drop=True)
train_info = df[['loss', 'acc', 'epoch']].dropna().reset_index(drop=True)
train_info = train_info.groupby('epoch').mean().reset_index(drop=True)
train_loss = train_info['loss']
train_acc = train_info['acc']
```

```python
train_acc.loc[199], val_acc.loc[199]
```

```python
# 画出acc和loss的变化
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(val_loss, label='val_loss')
plt.plot(train_loss, label='train_loss')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('loss')

plt.subplot(1, 2, 2)
plt.plot(val_acc, label='val_acc')
plt.plot(train_acc, label='train_acc')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('acc')
plt.title('acc')

plt.savefig(exp_name+'_loss_acc.png')
```
